## 11 -Get the number of reviews having positive sentiment and number of reviews having negative sentiment for the app 10 best foods for you and compare them.

```python
len(reviews[(reviews['App']=="10 Best Foods for You") & (reviews['Sentiment']=='Positive')])
```

```python
len(reviews[(reviews['App']=="10 Best Foods for You") & (reviews['Sentiment']=='Negative')])
```

## 12- Which comments of ASUS SuperNote have sentiment polarity and sentiment subjectivity both as 1?

```python
reviews[(reviews['App']=="ASUS SuperNote") & (reviews['Sentiment_Polarity']==1) & (reviews['Sentiment_Subjectivity']==1)]
```
